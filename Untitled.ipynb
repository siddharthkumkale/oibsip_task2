```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV files
df= pd.read_csv('Unemployment1.csv')
```

```python
df[' Date'] = pd.to_datetime(df[' Date'])
df.head()
```

```python
# Filter data for the Covid-19 period
start_date = pd.to_datetime('2020-01-01')
end_date = pd.to_datetime('2022-12-31')
covid_df = df[(df[' Date'] >= start_date) & (df[' Date'] <= end_date)]
```

```python
#plotting line graph
plt.figure(figsize=(10, 6))
plt.plot(covid_df[' Date'], covid_df[' Estimated Unemployment Rate (%)'])
plt.xlabel(' Date')
plt.ylabel(' Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate during Covid-19')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Plot histogram
plt.hist((covid_df[' Date'], covid_df[' Estimated Unemployment Rate (%)']), bins=20, edgecolor='black')

# Add labels and title
plt.xlabel(' Date')
plt.ylabel(' Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate during Covid-19')

# Show the plot
plt.show()
```
